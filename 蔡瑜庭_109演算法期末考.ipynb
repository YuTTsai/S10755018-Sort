使用堆排序，

將最大值排序在堆的上面，

若堆的最上面較小，則將較大值插入在堆的最上方，

堆的最上面較大則不動。

最終將堆排好取得第K大的值。

在此排好由大到小為 {9,8,6,5,4,3,2}。最終取得K=2，第二大值為8。

```
#程式   #參考資料https://www.itread01.com/content/1544687467.html

def Mykkk(PPPare,heap):
    CCChil = 2*PPPare+1
    while CCChil<len(heap):
        if CCChil+1<len(heap) and heap[CCChil+1]<heap[CCChil]:
            CCChil = CCChil+1
        if heap[PPPare]<= heap[CCChil]:
            break
        heap[PPPare],heap[CCChil] = heap[CCChil],heap[PPPare]
        PPPare,CCChil = CCChil,2*CCChil+1
    return heap
    
def MykkkFind_Mykkk(array,k):
    if k > len(array):
        return None
    heap = array[:k]
    for i in range(k,-1,-1):
        Mykkk(i,heap)
    for j in range(k,len(array)):
        if array[j]>heap[0]:
            heap[0] = array[j]
            Mykkk(0,heap)
    return heap[0]

print("請輸入要獲取的第K大值:")      
ABC=int(input())

CDE=[]
print("請輸入7個值:")
for n in range(7):       #重複輸入數字6次
   c=CDE.append(int(input())) 

print("陣列為:") 
print(CDE)
print("\n\n K=",ABC,"第二大值為 :", MykkkFind_Mykkk(CDE,ABC)) 
#[5,3,8,2,9,4,6]
```

第二題： 找零錢問題
目前有 1,2,5,10 四種面值的硬幣，請算出有總共有幾種方式湊N元?


Exp：輸入N=32，輸出：各種湊出32元的組合



解題想法：
使用動態規劃解題。
將問題分解成小問題，不用再重複計算相同的小問題，(節省時間，是一種以空間換取時間的方法)再由這些小問題組成大問題，解決完需要解決的問題。

將硬幣值1,2,5,10四種面值的硬幣，放在陣列arr裡，因此陣列arr為[1,2,5,10]。

m 為計算硬幣有幾種面值，此題為4，有四種面值1,2,5,10。

n 為幾元，此題為32元，要知道能組成32的有幾種方法。

定義函數count，計算出有幾種組合，並回傳結果。

在建構表時須建構 n+1 row，
從最小的面值開始加到32元，計算總共有幾次。
例如:從最小的面值1開始加上去，加完不會重複的之後就往下一個面值開始加，直到所有面值(4種)都加完，且不能有重複的組合，像是不能出現2種都是32個1組成32元的方法。

最後計算出以面值1,2,5,10元硬幣組成32源的方法共有"114種"。




```
#程式   

#參考資料：https://www.geeksforgeeks.org/python-program-for-coin-change/

# Dynamic Programming Python implementation of Coin  
# Change problem 
def count(S, m, n): 
    # We need n+1 rows as the table is constructed  
    # in bottom up manner using the base case 0 value 
    # case (n = 0) 
    table = [[0 for x in range(m)] for x in range(n+1)] 
  
    # Fill the entries for 0 value case (n = 0) 
    for i in range(m): 
        table[0][i] = 1
  
    # Fill rest of the table entries in bottom up manner 
    for i in range(1, n+1): 
        for j in range(m): 
  
            # Count of solutions including S[j] 
            x = table[i - S[j]][j] if i-S[j] >= 0 else 0
  
            # Count of solutions excluding S[j] 
            y = table[i][j-1] if j >= 1 else 0
  
            # total count 
            table[i][j] = x + y 
  
    return table[n][m-1] 
  
# Driver program to test above function 
arr = [1, 2, 5, 10] 
m = len(arr) 
n = 32
print('總共有:'+ str(count(arr, m, n)) +'種方式') 
  
# This code is contributed by Bhavya Jain 
```

#深度優先搜尋(DFS):

#廣度優先搜尋法(BFS):

以某個點當作起始點開始探索，往下探索，再將相鄰的那一層都探索完畢後，再往下一層探索，直到所有的點都探索完畢。

1.先從點1開始探索

2.接著從點2、點3中選擇一個開始探索，在此選擇點2。

3.再探索相鄰的點3。

4.該層已無相鄰未探索的點，因此往下一層探索，再下一層中，從相鄰的點4、點5、點6中選擇一點開始探索一點開始探索，在此選擇點4。

5.探索完點4之後，再探索相鄰的點5、點6選擇一點探索，在此選擇點5。

6.探索完點5之後，再探索相鄰的點6。

7.該層已無相鄰未探索的點，因此往下一層探索，在下一層中，從相鄰的點8、點9中選擇一點開始探索，在此選擇點8。

8.探索完點8之後，再探索相鄰的點9。

9.該層已無相鄰未探索的點，因此往下一層探索，下一層中只有點7，因此探索點7。

所有的節點皆已探索完畢。
探索順序為:1,2,3,4,5,6,8,9,7




```
#程式　　

#參考資料：https://misavo.com/blog/%E9%99%B3%E9%8D%BE%E8%AA%A0/%E6%9B%B8%E7%B1%8D/%E4%BA%BA%E5%B7%A5%E6%99%BA%E6%85%A7/04-%E5%9C%96%E5%BD%A2%E6%90%9C%E5%B0%8B/B-%E5%AF%A6%E4%BD%9C%EF%BC%9A%E6%B7%B1%E5%BA%A6%E8%88%87%E5%BB%A3%E5%BA%A6%E6%90%9C%E5%B0%8B%E6%B3%95

#將ＡＢＣＤＥＦＧＨＩ改為１２３４５６７８９　

def enqueue(a, o):
    a.insert(0, o)

def dequeue(a):
    return a.pop()

g = { #  graph: 被搜尋的網路
    '1': {'n':['2','3'], 'v':0}, #A  #  n: neighbor (鄰居), v: visited (是否被訪問過)
    '2': {'n':['4','5'], 'v':0}, #B
    '3': {'n':['6'], 'v':0},   #C
    '4': {'n':[], 'v':0},    #D
    '5': {'n':['8'], 'v':0},   #E
    '6': {'n':['9'], 'v':0},   #F
    '7': {'n':[], 'v':0},     #G
    '8': {'n':['7'], 'v':0},   #H
    '9': {'n':[], 'v':0}     #I
}


def init(g): #  初始化、設定 visited 為 0
    for i in g:
        g[i]['v'] = 0

def dfs(g, node): #  深度優先搜尋
    if g[node]['v']!=0:           #  如果已訪問過，就不再訪問
        return
    print(node, '=> ', end = '')  #  否則、印出節點
    g[node]['v'] = 1              #    並設定為已訪問
    neighbors = g[node]['n']      # 取出鄰居節點
    for n in neighbors:           #  對於每個鄰居
        dfs(g, n)                 #    逐一進行訪問

queue=['1'] #  BFS 用的 queue, 起始點為 1。

def bfs(g, q): #  廣度優先搜尋
    if len(q)==0:                 #  如果 queue 已空，則返回。
        return
    node = dequeue(q)             #  否則、取出 queue 的第一個節點。
    if g[node]['v'] == 0:         #  如果該節點尚未拜訪過。
        g[node]['v'] = 1          #    標示為已拜訪
    else:                         #  否則 (已訪問過)
        return                    #    不繼續搜尋，直接返回。
    print(node, '=> ', end = '')  #  印出節點
    neighbors = g[node]['n']      #  取出鄰居。
    for n in neighbors: #  對於每個鄰居
        if not g[n]['v']:         #  假如該鄰居還沒被拜訪過
            enqueue(q, n)         #    就放入 queue 中
    bfs(g, q)

print('dfs深度優先搜尋法:', end = '')
init(g)
dfs(g, '1') # 呼叫深度優先搜尋。
print('所有點皆已探索完畢')
print('')

print('bfs廣度優先搜尋法:', end = '')
init(g)
bfs(g, queue) # 呼叫廣度優先搜尋。
print('所有點皆已探索完畢')
print('')
```
